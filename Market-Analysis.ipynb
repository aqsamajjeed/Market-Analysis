```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

**When was the global search for 'workout' at its peak?**

```python
# Start coding here
import pandas as pd

workout = pd.read_csv("data/workout.csv")
three_keywords = pd.read_csv("data/three_keywords.csv")
workout_geo = pd.read_csv("data/workout_geo.csv")
three_keywords_geo = pd.read_csv("data/three_keywords_geo.csv")

# Find the month with the highest 'workout_worldwide' value
peak_month = workout.loc[workout['workout_worldwide'].idxmax(), 'month']
# The idxmax() method in Pandas returns the index of the first occurrence of the maximum value along a specified axis in a DataFrame.

# Extract the year from the peak month
year_str = peak_month[:4]
print(f"In {year_str}, the global search for the keyword workout was at its peak.")
```

**Of the keywords available, what was the most popular during the covid pandemic, and what is the most popular now?**

**What country has the highest interest for workouts among the following: United States, Australia, or Japan?**

```python
# Filter the workout_geo dataframe for the countries of interest
countries_of_interest = ["United States", "Australia", "Japan"]
workout_geo = workout_geo[workout_geo["country"].isin(countries_of_interest)]

# Find the country with the highest interest for workouts
top_country = workout_geo.loc[workout_geo["workout_2018_2023"].idxmax(),"country"]

#print
print(f"{top_country} has the highest interest in workouts")
```
